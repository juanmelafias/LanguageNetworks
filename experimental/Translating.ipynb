```python
import pandas as pd
filelist = os.listdir('./files/inflected/dictionaries/')
languagelist = [file.split('.')[0] for file in filelist]
tokens = []
for language in languagelist:

    df = pd.read_csv(f'files/inflected/dataframes/{language}.csv').rename(columns={"Unnamed: 0": "unique_id"})
    sumcount= df['count'].sum()
    tokens.append(sumcount)
```

```python
tokens
```

```python
caca = pd.read_csv('files/lemmatized/stats.csv').drop(axis = 1,labels=['Unnamed: 0']).transpose()
```

```python
for index,lang in enumerate(languagelist):
    bv = caca[lang].to_list().append(tokens[index])
    caca[lang] = bv
```

```python
caca
```

```python
caca 
```

```python
caca['tokens'] = tokens
```

```python
caca['language'] = caca.index
caca.columns = ['nodes GCC', 'edges', 'tokens', 'language']
caca
```

```python
caca = caca.reset_index(level=0)
caca
```

```python
caca.to_csv('files/lemmatized/stats.csv')
```
