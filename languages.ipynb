```python
def relevant_names(x):
    lista = []
    for i,name in enumerate(x):
        if name[-1] == 'u':
            lista.append(name)
    
    return lista

```

```python
df['names'] = df.directories.apply(lambda x: x.replace('/',''))
df['corpus'] = df.names.apply(lambda x: x.split('-')[1])
df['language'] = df.names.apply(lambda x: x.split('-')[0].replace('UD_','').replace('_',' '))
df['filenames'] = df.directories.apply(lambda x: os.listdir(f'./UsefulData/ud-treebanks/ud-treebanks/{x}'))
df['relevant'] = df.filenames.apply(relevant_names)
```

```python
def get_lines(filePath):
    fIn = open(filePath,encoding = "utf8"); 
    allLines = fIn.read().split('\n'); 
    fIn.close(); 
    numlines = len(allLines)
    return numlines
   
```

```python
def get_cols(filePath):
    
    fIn = open(filePath,encoding = "utf8"); 
    allLines = fIn.read().split('\n'); 
    fIn.close(); 
    line = True
    i = 0
    while line:
        if allLines[i] == '':
            i = i+1
        elif allLines[i][0] == '#':
            i = i+1
        else:
            line = False
    numcols = len(allLines[i].split('\t'))
    return numcols
```

```python
df['prueba'] = df['relevant'].apply(lambda x: [df['directories']+f for f in x])
```

```python
for i in range(len(df)):
    df['prueba'][i] = ['UsefulData/ud-treebanks/ud-treebanks/'+df['directories'][i]+f for f in df['relevant'][i]]
df['prueba'][0]
```

```python
df['num_lines'] = df.prueba.apply(lambda x: [get_lines(f) for f in x])
df['num_cols'] = df.prueba.apply(lambda x: [get_cols(f) for f in x])
df
```

```python
for row in range(0,len(df)):
    corpus = df['corpus'][row]
    language = df['language'][row]
    for index,file in enumerate(df['prueba'][row]):
        prueba = file
        num_lines = df['num_lines'][row][index]
        num_cols = df['num_cols'][row][index]
        new_row = {'corpus':corpus,'language':language,'prueba':prueba, 'num_lines':num_lines, 'num_cols':num_cols}
        individual = individual.append(new_row, ignore_index=True)

        
```

```python
individual = individual.rename(columns = {'prueba':'location'})
```

```python
list_languages = individual.language.unique()
routes = []
individual['max'] = 'No'
for language in list_languages:
    individual_lang = individual[individual['language']==language]
    print(individual_lang)
    route = individual_lang[individual_lang['num_lines']== max(individual_lang['num_lines'])]['location'].tolist()[0]
    lines = individual[individual['location']== route]['num_lines'].tolist()[0]
    lang = individual[individual['location']== route]['language'].tolist()[0]
    cols = individual[individual['location']== route]['num_cols'].tolist()[0]
    minidict = {'route':route, 'num_lines':lines, 'language':lang, 'num_cols':cols}
    routes.append(minidict)
len(routes),len(list_languages)
```

```python
valid_files = pd.DataFrame(routes)
valid_files.to_csv('rutascorpus.csv')
```
